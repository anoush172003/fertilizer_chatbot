```python
import pandas as pd
import numpy as np
f1=pd.read_csv(r'C:\Users\User\Desktop\fertilizers\Untitled Folder\UdupiSoil-Data-_1_.csv')
f2=pd.read_csv(r'C:\Users\User\Desktop\fertilizers\Untitled Folder\crop_recommendation.csv')
f3=pd.read_csv(r'C:\Users\User\Desktop\fertilizers\Untitled Folder\Fertilizer Prediction.csv')
```

```python
f1.columns = f1.columns.str.strip()
f2.columns = f2.columns.str.strip()
f3.columns = f3.columns.str.strip()
```

```python
f2['Temperature'] = f2['Temperature'].apply(pd.to_numeric, errors='coerce').round()
f2['Humidity'] = f2['Humidity'].apply(pd.to_numeric, errors='coerce').round()
f3['Fertilizer Name']=f3['Fertilizer Name'].astype(str)
```

```python
merged_ds=pd.merge(f3,f2,on=['N','P','K','Temperature','Humidity'],how='left')
```
